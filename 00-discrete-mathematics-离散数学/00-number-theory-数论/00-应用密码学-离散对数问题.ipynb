```python
import math
# æ¬§æ‹‰å‡½æ•°ï¼šæ ¹æ®å®šä¹‰
def phi(n):
    ct = 0
    for i in range(1, n):
        if math.gcd(i, n) == 1:
            ct += 1
    return ct

# tests
numbers = [7, 10, 8]
answers = [6, 4, 4]
for n, a in zip(numbers, answers):
    print('phi({}) = {}, ans = {}'.format(n, phi(n), a))
```

```python
# æ±‚ä¸å«è‡ªèº«çš„å› æ•°é›†åˆ
def factors(n):
    fac = list()
    for i in range(1, n):
        if n % i == 0:
            fac.append(i)
    return fac
          
# æŒ‰å®šä¹‰è®¡ç®—æœ€å°åŸæ ¹
def min_primitive_root(n):
    phin = phi(n)
    factors_ = factors(phin)
    for a in range(2, n):
        if math.gcd(a, n) != 1: # äº’è´¨
            continue
        flag = True
        for d in factors_:
            x = a ** d % n
            if x == 1:
                flag = False
                break
        # æ‰€æœ‰d | phi(n), n != phi(n), a^d (mod n) != 1
        if flag:
            return a # ç”±åŸæ ¹çš„è®¡ç®—ï¼ˆ1ï¼‰ä¸€å®šæœ‰ï¼ša^phi(n) (mod n) = 1
    return None

# æŒ‰åŸæ ¹çš„è®¡ç®—ï¼ˆ1ï¼‰è®¡ç®—æœ€å°åŸæ ¹
def min_primitive_root1(n):
    phin = phi(n)
    factors_ = integer_factorization(phin)
    # é¢„å¤„ç† phin/pi
    factors_ = [int(phin / fac) for fac in factors_]
    for a in range(2, n):
        if math.gcd(a, n) != 1: # äº’è´¨
            continue
        flag = True
        for d in factors_:
            x = a ** d % n
            if x == 1:
                flag = False
                break
        # æ‰€æœ‰d | phi(n), n != phi(n), a^d (mod n) != 1
        if flag:
            return a # ç”±æ¬§æ‹‰å®šç†ä¸€å®šæœ‰ï¼ša^phi(n) (mod n) = 1
    return None

numbers = list(range(2, 32))
numbers.append(41)
numbers.append(761)
answers = [None, 2, 3, 2, 5, 3, None, 2, 3,\
           2, None, 2, 3, None, None, 3, 5, 2, None,\
           None, 7, 5, None, 2, 7, 2, None, 2, None, 3, 6, 6]
ok_ct = 0
for n, expected in zip(numbers, answers):
    phin = phi(n)
    a = min_primitive_root(n)
    a1 = min_primitive_root1(n)
    assert(a == a1)
    state = 'Bad'
    if a == expected:
        state = 'Ok'
        ok_ct += 1
    print('n={}: phi = {}, a = {}, expected = {}, state = {}'\
        .format(n, phin, a, expected, state))
print('OK: {:.2f}%'.format(ok_ct * 1.0/ len(numbers) * 100))
```

## 5.2 [è´¨å› æ•°åˆ†è§£](https://baike.baidu.com/item/%E5%88%86%E8%A7%A3%E8%B4%A8%E5%9B%A0%E6%95%B0/2253749)
æŠŠä¸€ä¸ªåˆæ•°åˆ†è§£æˆè‹¥å¹²ä¸ªè´¨å› æ•°çš„ä¹˜ç§¯çš„å½¢å¼ï¼Œå³æ±‚è´¨å› æ•°çš„è¿‡ç¨‹å«åšåˆ†è§£è´¨å› æ•°ã€‚ åˆ†è§£è´¨å› æ•°åªé’ˆå¯¹åˆæ•°ã€‚\
ä¾‹å¦‚ $24 \; = \; 2 ^ 3 * 3$ã€$2 = 2^1$\
æ¨è®ºï¼šå¯¹ä¸€ä¸ªæ­£æ•´æ•° $n$ æ¥è¯´ï¼Œå¦‚æœå®ƒå­˜åœ¨ $[2,n]$ èŒƒå›´å†…çš„è´¨å› å­ï¼Œè¦ä¹ˆè¿™äº›è´¨å› å­å…¨éƒ¨å°äºç­‰äº $\sqrt n$ ï¼Œè¦ä¹ˆåªå­˜åœ¨ä¸€ä¸ªå¤§äº $\sqrt n$ çš„è´¨å› å­ï¼Œè€Œå…¶ä½™è´¨å› å­å…¨éƒ¨å°äºç­‰äº $\sqrt n$ \
è¯æ˜ï¼ˆåæ­£æ³•ï¼‰ï¼šå‡è®¾å­˜åœ¨å¤šäºä¸€ä¸ªå¤§äº $\sqrt n$ çš„è´¨å› æ•°ï¼Œå…¶ä¸­çš„ä»»æ„ä¸¤ä¸ªä¸º$a,b > \sqrt n$ï¼Œåˆ™æœ‰è´¨å› æ•°åˆ†è§£$n \;= \; a ^ x b^yR > ({\sqrt n})^2R = nR >= n \Rightarrow n > n$,ä¸$n\;=\;n$çŸ›ç›¾ï¼Œæ•…å‡è®¾ä¸æˆç«‹ã€‚å…¶ä¸­$R$è¡¨ç¤ºå‰©ä½™çš„è´¨å› æ•°åˆ†è§£ï¼Œä¸” $R \ge 1$ 

**æœ´ç´ ç®—æ³•**å®ç°å¦‚ä¸‹ï¼š

```python
import math
# O(sqrt(n))
def integer_factorization(n):
    r = math.ceil(math.sqrt(n))
    fac = []
    # step 1: [1, r]
    for i in range(2, r +1):
        if n % i == 0:
            fac.append(i)
            while n % i == 0:
                n = n/i
    # step 2: [r + 1, n]
    if n > 1:
        fac.append(n)
    return fac

# cell tests
numbers = [1, 2, 3, 5, 8, 9, 11, 12, 24, 63, 1024, 8232, 760, 100000007 * 100000037]
excepted = [[], [2], [3], [5], [2], [3], [11], [2, 3], [2, 3], [3, 7], [2], [2, 3, 7], [2, 5, 19], [100000007, 100000037]]
ok_ct = 0
for n, e in zip(numbers, excepted):
    result = integer_factorization(n)
    ok = (e == result)
    print('integer_factorization({}): {}, except={}, ok = {}'\
        .format(n , result, e, ok))
    if ok:
        ok_ct += 1
print('OK: {:.2f}%'.format(ok_ct * 1.0 / len(numbers) * 100))
```
